```python
import cv2 as cv
 # cap name ka aik Variable Create kia hai
cap = cv.VideoCapture(0)
# jb hum Video Record aur save karta hain to yeh by default kuch Fourcc codes use karta hain 
fourcc = cv.VideoWriter_fourcc(*'XVID')    

''' out name ka aik variable create kia hai aur aik Class use ki hai Video Writer ka name sa 
jo 4 Arguments leta hai 
First(jo Video video save karni hai us ka name)
Second(fourcc code ko jis variable ma ave kia hai wo)
Third(Frame per second dena hai)
Fourth(Us Video ka size)'''

out = cv.VideoWriter("Output.avi", fourcc, 30.0, (640,480)) 

'''While Loop Create kia hai is ma hum na method use kia hai isOpened ka name sa jo hamain 
Ture Return kare ga agr Camera Open ho Aur False return kare ga agr na open ho'''
while(cap.isOpened()):

    ret, frame = cap.read()

    if ret == True:

        # Frame ki Height Aur Widht Dekhna ka lia yeh 2 Lines of Code Likha hai
        print(cap.get(cv.CAP_PROP_FRAME_WIDTH))
        print(cap.get(cv.CAP_PROP_FRAME_HEIGHT))

        
        # Yeh Line of Code Video Ko record Ya Write Karna ka lia hai
        out.write(frame)
#         Video ka color ko Gray ma Change Karna ka lia
#         gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)

        cv.imshow("Camera", frame)

        # Video ko Close krna ka lia koi charcter Chaey hum na q use kia hai usa close karna ka lia
        if cv.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
cap.release()
out.release()
cv.destroyAllWindows()
```
